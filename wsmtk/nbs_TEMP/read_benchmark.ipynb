```python
def load1(x,c):
    
    ysize = c // 2400
    
    for c1 in range(0,2400**2,c): 
        
        yoff = c1 // 2400
        
        ds = gdal.Open(x)
        sds_o = gdal.Open(ds.GetSubDatasets()[0][0])
    
        arr = sds_o.ReadAsArray(0,yoff,2400,ysize)
    
        ds = None
        sds_o = None
    
def load2(x,c):
    tt = []
    ysize = c//2400
    ds = gdal.Open(x)
    sds_o = gdal.Open(ds.GetSubDatasets()[0][0])
    
    for ii in range(10):
        
        t1 = time.time()
        
        for c1 in range(0,2400**2,c):
            
    
            yoff = c1 // 2400
        
            
    
            arr = sds_o.ReadAsArray(0,yoff,2400,ysize)
            
            print(arr)
            
    
        tt.append(time.time()- t1)
    
    ds = None
    sds_o = None
    
    return(tt)
```
